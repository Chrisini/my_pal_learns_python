```python
import pandas as pd
import glob

url = 'https://raw.githubusercontent.com/GMU-CherryBlossomCompetition/peak-bloom-prediction/main/data/japan.csv'
df = pd.read_csv(url, delimiter=",") 

print(df.head(5))

df = df.drop(columns=["long", "lat", "alt", "bloom_date"]).rename(columns={"bloom_doy":"doy_cherry_peak_bloom"})

#create unique list of names
unique_places = df.location.unique()

#create a data frame dictionary to store your data frames
DataFrameDict = {elem : pd.DataFrame() for elem in unique_places}

for key in DataFrameDict.keys():
    # DataFrameDict[key] = data[:][data.Names == key]
    print("")
    print(df[:][df.location == key].head(2))
    print(df[:][df.location == key].iloc[-2:])
    this_df = df[:][df.location == key]
    

    loc = (key).replace("/", "_") # this_df["location"][0]
    
    this_df = this_df.drop(columns=["location"])
    
    this_df.to_csv(f"../../datasets/csv/cherry/doy_cherry_peak_bloom_{loc}.csv", index=False)
    

```
